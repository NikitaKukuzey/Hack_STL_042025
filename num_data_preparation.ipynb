```python
df = pd.read_csv("/kaggle/input/vk-predict-hackathon/train.csv")
```

```python
df
```

```python
df.describe()
```

Видим, что есть столбцы с константными значениями. Избавимся от них

```python
cols_to_drop = df.columns[df.nunique() == 1]
cols_to_drop
```

```python
df = df.drop(columns=cols_to_drop)
```

```python
types_df = df.dtypes.reset_index()
types_df.columns = ['Column', 'Type']
type_summary = types_df.groupby('Type').count()
print(type_summary)
```

Рассмотрим целочисленные столбцы

```python
int_df = df.select_dtypes(include=['int64'])
int_df = int_df.drop(columns=['index'])
int_df
```

```python
value_counts_df = pd.DataFrame({
    col: int_df[col].value_counts()
    for col in int_df.columns
})

print(value_counts_df)
```

```python
columns_of_interest = int_df.columns
correlation_matrix = df[columns_of_interest].corr()

print("Таблица корреляции:")
print(correlation_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Строим тепловую карту
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Тепловая карта корреляции")
plt.show()
```

Теперь рассмотрим float-столбцы

```python
f_df = df.select_dtypes(include=['float64'])
f_df
```

Выберем столбцы с небольшим количеством значений

```python
lf_columns = f_df.columns[f_df.nunique() > 10]
bin_columns = f_df.columns[f_df.nunique() == 2]
cat_columns = f_df.columns[(f_df.nunique() > 2) & (f_df.nunique() <= 10)]
```

Оставляем bin_columns как есть, потом объединим их с int_columns

К cat_columns применим разбиение на несколько признаков

```python
cat_df = f_df[cat_columns]
```

```python
cat_df
```

```python
dum_df = pd.get_dummies(cat_df, columns=cat_columns)
```

```python
dum_df
```

В некоторых столбцах преобладает False. Удалим их

```python
true_counts = dum_df.astype(bool).sum(axis=0)
dum_df = dum_df[true_counts[true_counts >= 1000].index]
```

```python
dum_df
```

Теперь обработаем float столбцы

```python
lf_df = f_df[lf_columns]
```

```python
lf_df
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
scaled_data = scaler.fit_transform(lf_df[lf_columns])
lf_df = pd.DataFrame(scaled_data, columns=lf_columns)
```

```python
lf_df
```

Теперь собираем датафреймы в один

```python
int_df = int_df.drop(columns=['target'])
```

```python
y_df = df[["index", "target"]]
```

Обработаем датафрейм со столбцами с двумя уникальными float-значениями

```python
bin_df = f_df[bin_columns]
```

```python
for col in bin_columns:
    unique_values = bin_df[col].unique()
    value_map = {unique_values[0]: 1.0, unique_values[1]: 0.0}
    bin_df.loc[:, col] = bin_df[col].map(value_map)
```

```python
bin_df
```

```python
result = pd.concat([y_df, int_df, bin_df, dum_df, lf_df], axis=1)
```

```python
result
```

```python
result.to_csv('/kaggle/working/ptrain_v1.csv', encoding='utf-8')
```
